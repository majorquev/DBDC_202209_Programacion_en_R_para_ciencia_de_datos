## Producto interno bruto

El producto interno bruto (PIB) es una medida monetaria del valor de mercado de todos los bienes y servicios finales producidos y vendidos (no revendidos) en un período de tiempo específico por países. Debido a su naturaleza compleja y subjetiva, esta medida a menudo se revisa antes de ser considerada un indicador confiable. El PIB (nominal) per cápita, sin embargo, no refleja las diferencias en el costo de vida y las tasas de inflación de los países; por lo tanto, usar una base de PIB per cápita en paridad de poder adquisitivo (PPA) puede ser más útil cuando se comparan los niveles de vida entre naciones, mientras que el PIB nominal es más útil para comparar economías nacionales en el mercado internacional. El PIB total también se puede desglosar en la contribución de cada industria o sector de la economía. La relación entre el PIB y la población total de la región es el PIB per cápita (también llamado Nivel de vida medio).

Fuente: https://en.wikipedia.org/wiki/Gross_domestic_product

### Pregunta 3 - Construcción Crecimiento PIB acumulado base 100

En base a la variable gdp_growth podemos construir el Crecimiento PIB acumulado con base 100 (el primer valor que tomará la serie del PIB será 100). Para esto debemos hacer uso de la siguiente fórmula


\begin{align}
PIB_{100}= 100 \cdot \prod_{i=1}^n \left(\frac{gdp_{growth,i}}{100} + 1\right)
\end{align}

es decir que si el pais *imaginario* tiene los años 2018, 2019, 2020 y 2021 un crecimiento del PIB de 1.5%, 4.8%, -5.1% y -3.9% respectivamente, entonces su crecimiento acumulado desde 2018 será
\begin{align}
    2018 → 100 \cdot ( \frac{1.5}{100}+1 )= 101.50 \\[1em]
    2019 → 100 \cdot ( \frac{1.5}{100}+1 )\cdot( \frac{4.8}{100}+1 )= 106.37 \\[1em]
    2020 → 100 \cdot ( \frac{1.5}{100}+1 )\cdot( \frac{4.8}{100}+1 ) \cdot( \frac{-5.1}{100}+1 )= 100.95 \\[1em]
    2021 → 100 \cdot ( \frac{1.5}{100}+1 )\cdot( \frac{4.8}{100}+1 ) \cdot( \frac{-5.1}{100}+1 )\cdot( \frac{-3.9}{100}+1 )= 97.01
\end{align}

En R, usted puede realizar esto usando la función cumprod()



```r
pib = c(1.5,4.8,-5.1,-3.9)

pib_historico = 100*cumprod(pib/100+1)

pib_historico
```
